# HW01 — NumPy basics

```python
# Импорт NumPy
import numpy as np
```

## Создание массивов

```python
# Массив из обычного списка
a = np.array([1, 2, 3, 4])
a
```

```python
# Массив из zeros
z = np.zeros((2, 3))
z
```

```python
# Массив с arange
b = np.arange(0, 10, 2)
b
```

## Атрибуты массива

```python

a.shape, a.dtype
```

## Индексация и срезы

```python
# Один элемент
elem = b[2]
elem
```

```python
# Срез подмассива
sub = b[1:4]
sub
```

## Векторные операции

```python
# Поэлементное сложение
c = a[:4] + np.array([10, 10, 10, 10])
c
```

```python
# Среднее значение массива
mean_val = a.mean()
mean_val
```
